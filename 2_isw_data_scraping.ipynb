```python
def remove_names_and_dates(page_html_text):
    parsed_html = BeautifulSoup(page_html_text)
    p_lines = parsed_html.findAll('p')
    
    min_sentences_word_count = 20
    p_index=0
    
    for p_line in p_lines:
        strong_lines = p_line.findAll('strong')
        if not strong_lines:
            p_index += 1
            continue
            
        for s in strong_lines:
            if len(s.text.split(' '))>= min_sentences_word_count:
                break
        else:
            p_index +=1
            continue
        break
    for i in range(0, p_index):
        page_html_text = page_html_text.replace(str(p_lines[i]), '')
    return page_html_text
```

```python
df['main_html_v2'] = df['main_html'].apply(lambda x: remove_names_and_dates(x))
```

```python
df['main_html_v3'] = df['main_html_v2'].apply(lambda x: remove_names_and_dates(x))
```

```python
df['main_html_v4'] = df['main_html_v3'].apply(lambda x: remove_names_and_dates(x))
```

```python
import re

pattern = '\[(\d+)\]'

df['main_html_v5'] = df['main_html_v4'].apply(lambda x: re.sub(pattern, '', x))
```

```python
df['main_html_v5'] = df['main_html_v4'].apply(lambda x: BeautifulSoup(x).text)
df
```

```python
df['main_html_v6'] = df['main_html_v5'].apply(lambda x: re.sub(r'http(\S+.*\s)','',x))
df['main_html_v7'] = df['main_html_v6'].apply(lambda x: re.sub(r'(o2022|o2023|2022|2023)','',x))
```

```python
def del_clickHere(df):
    divider = "in this report."
    for i in range(df.shape[0]):
        if divider in df.iloc[i,7]:
            temp_str = df.iloc[i,7]
            temp_list = temp_str.split(divider)
            df.iloc[i,7] = temp_list[1]
```

```python
del_clickHere(df)
df
```

```python
def del_xa0(df):
    for i in range(df.shape[0]):
        df.iloc[i,7] = df.iloc[i,7].replace('\xa0','')
```

```python
del_xa0(df)
```

```python
def del_brackets(df):
    for i in range(df.shape[0]):
        temp_list = df.iloc[i,7].split('\n[1]')
        df.iloc[i,7] = temp_list[0]
```

```python
del_brackets(df)
df
```
